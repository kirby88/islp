# Exercise 7

The response $Y$ is a categorical variable that can be either Yes or No. And $X$ the predictor that is last year's profit in percentage.

When $Y=$ Yes, then $\bar{X} = 10$
When $Y=$ No, then $\bar{X} = 0$

$\sigma^2 = 36$ and 80% of the companies issued dividends ($Y=$ Yes)

We know that $X \sim \mathcal{N}(\mu, \sigma)$.
So the probability that a company will issue a dividend this year given it's percentage profit was $X=4$ is:

$$
\begin{align*}
\displaystyle 
p(x) &= \frac{\pi_k f_k(x)}{\sum \pi_l f_l(x)}
\end{align*}
$$

In our case, we have two classes (Yes and No). And we know the prior probabilities, so $\pi_{Yes} = 0.8$ and $\pi_{No} = 0.2$.
$$
\begin{align*}
\displaystyle 
f_{Yes}(x) &= \frac{1}{\sqrt{2\pi}\sigma_{Yes}}\exp(-\frac{1}{2\sigma^2_{Yes}}(x-\mu_{Yes})^2)
\end{align*}
$$

We already know $\mu_{Yes}$ which the average profit percentage when companie is given profit, and is 10.  
$\sigma^2 = 36$ so $\sigma = 6$, that we inject in the previous equation:

$$
\begin{align*}
\displaystyle 
f_{Yes}(x) &= \frac{1}{\sqrt{2\pi}*6}\exp(-\frac{1}{2*36}(x-10)^2)
\end{align*}
$$

So for a percentage profit of $x=4$, we get $f_{Yes}(4) = 0.04032845408$

We then do the same for $f_{NO}$

$$
\begin{align*}
\displaystyle 
f_{No}(x) &= \frac{1}{\sqrt{2\pi}\sigma_{No}}\exp(-\frac{1}{2\sigma^2_{No}}(x-\mu_{No})^2)\\
 &= \frac{1}{\sqrt{2\pi}*6}\exp(-\frac{1}{2*36}(x)^2)
\end{align*}
$$

So for a percentage profit of $x=4$, we get $f_{No}(4) = 0.05324133425$

Now to get the final answer, we just inject these values in the first equation

$$
\begin{align*}
\displaystyle 
p(Y=Yes | X=x) &= \frac{\pi_{Yes}f_{Yes}(x)}{\pi_{Yes}f_{Yes}(x)+\pi_{No}f_{No}(x)}\\
&= \frac{0.8*0.04032845408}{0.8*0.04032845408+0.2*0.05324133425}\\
&= 0.75185245328
\end{align*}
$$

So at the end the chances for a company that has a percentage profit of 4% to give dividends this year is 75.2%
