## a.

```python
def Power():
    print(f"2^3 = {2**3}")

Power()
```

## b.

```python
def Power2(x, a):
    print(f"{x}^{a} = {x**a:,}") # :, means that we want to format the number as a string with comma for better readability

Power2(3,8)
```

## c.

```python
Power2(10,3)
Power2(8,17)
Power2(131,3)
```

## d.

```python
def Power3(x, a):
    return x**a

result = Power3(3,8)
print(f"3^8 = {result:,}")
```

## e.

```python
from matplotlib.pyplot import subplots
import numpy as np

fig_month, ax_month = subplots(figsize=(8,8))
x = np.arange(1, 11)
ax_month.plot(x, Power3(x,2), marker="o", ms=10)
# ax_month.set_yscale("log")
# ax_month.set_xscale("log")
ax_month.set_xlabel("x", fontsize=20)
ax_month.set_ylabel("x^2", fontsize=20);
```

## f.

```python
def PlotPower(x, a):
    _, ax_month = subplots(figsize=(8,8))
    ax_month.set_yscale("log")
    ax_month.plot(x, Power3(x,a), marker="o", ms=10)
    ax_month.set_xlabel("x", fontsize=20)
    ax_month.set_ylabel(f"x^{a}", fontsize=20);

PlotPower(np.arange(1, 11), 3)
```
